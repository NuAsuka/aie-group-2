# Копии и представления

```python
a = np.arange(10)
a
```

```python
b = a
```

```python
b
```

```python
b is a # b и a ссылаются на один объект
```

```python
b.shape = (5, 2)
```

```python
a # изменение объекта меняет значения b и a одновременно
```

```python
c = a.view()
c
```

```python
c is a
```

```python
a.shape = (2, 5)
c
```

```python
c[0][0] = 100
c[1,0] = 1000
```

```python
a
```

```python
c # view даёт создавать разные представления, но значения остаются общие (ячейки памяти общие)
```

```python
d = a.copy()
```

```python
d is a
```

```python
d[0,0] = 999
a
```

```python
d # теперь всё хранится в разных ячейках памяти
```
